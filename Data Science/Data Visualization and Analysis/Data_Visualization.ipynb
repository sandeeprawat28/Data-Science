Basic Data Visualization

```python
# import the dependencies
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns
sns.set()
```

```python
# getting random values for y
y = 200 + np.random.randn(100)
print(y)
type(y)
```

```python
x = [x for x in range(len(y))]  # list containing 0 - 99
print(x)
type(x)
```

```python
# plot x and y
plt.plot(x,y,'-')
plt.title('Data Visualization Example')
plt.show()
```

```python
# fill the area under the line
plt.fill_between(x,y,195,where=(y>195),facecolor='g',alpha=0.5)
plt.title('Data Visualization Example')
plt.show()
```
