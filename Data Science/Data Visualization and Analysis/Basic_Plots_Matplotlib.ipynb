```python
# import the dependencies
import numpy as np
import matplotlib.pyplot as plt  # plotting library
import math
```

```python
# getting values
a = np.arange(0,math.pi*2,0.04)  # pi*2 = 360
b = np.sin(a)

print(a)
print(b)
```

```python
# plot a nnd b
plt.plot(a,b)
plt.xlabel('Angle')
plt.ylabel('Sing Value')
plt.title('Sine Function')
plt.grid(True) # add grid in graph
plt.show()
```

# Important Plots in Data Visualization

1. Scatter Plot
2. Bar Graph
3. Pie Chart

1. Scatter Plot

```python
# Scatter Plot
x = 100 + np.random.randn(50)
y = 100 + np.random.randn(50)

plt.scatter(x,y)
plt.xlabel('X-Value')
plt.ylabel('Y-Value')
plt.title('Scatter Plot')
plt.show()
```

2. Bar Graph

```python
# Bar Graph
x = ['Jan','Feb','Mar','Apr','May']
y = [100,200,156,234,850]

plt.bar(x,y)
plt.title('Phone Sale')
plt.xlabel('Months')
plt.ylabel('Number of Phones Sold')
plt.show()
```

3. Pie Chart

```python
# Pie Chart
sports = ['Cricket','Football','Hockey','Tennis','Basketball']
players = [964,266,852,860,746]

plt.pie(players, labels=sports, autopct='%.2f%%')
plt.show()
```
