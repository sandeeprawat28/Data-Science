# NaN - Not a Number

```python
# importing the dependencies
import numpy as np
import pandas as pd
```

```python
# Loading the dataset
dataset = pd.read_csv('/content/Melbourne_housing_FULL.csv')
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
# Number of missing values for each column
dataset.isnull().sum()
```

1. Filling BuildingArea with Mean value

```python
dataset['BuildingArea'].fillna(dataset['BuildingArea'].mean(), inplace=True)
```

```python
dataset.head()
```

2. Filling the Price value with Median

```python
dataset['Price'].fillna(dataset['Price'].median(), inplace=True)
```
