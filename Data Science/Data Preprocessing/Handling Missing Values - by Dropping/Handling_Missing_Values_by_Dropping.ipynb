```python
# importing the dependencies
import numpy as np
import pandas as pd
```

```python
# Loading the dataset
dataset = pd.read_csv('/content/Melbourne_housing_FULL.csv')
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
#checking which values are missing in the first 5 rows
dataset.isnull().head()  # Missing = True,
```

```python
# Number of missing values for each column
dataset.isnull().sum()
```

```python
# Identify the rows where the CouncilArea values is missing
dataset[dataset.CouncilArea.isnull()] 
```

```python
# Dropping the rows even if contains one missing value
dataset.dropna(how='any').shape  # use inplace = true for changing origina data
```

```python
# Dropping the rows even if contains all missing value
dataset.dropna(how='all').shape
```

```python
dataset.shape
```

```python
# Dropping the columns containing missing values
dataset.dropna(axis=1)
```
