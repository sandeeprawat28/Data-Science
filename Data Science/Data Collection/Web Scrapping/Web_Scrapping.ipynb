```python
# importing the dependencies
from bs4 import BeautifulSoup  #bs helps to get data from web pages
import requests  # used to send http request
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Rice_production_in_India'  # page url

request_response = requests.get(url)
soup = BeautifulSoup(request_response.text, 'html.parser')
```

```python
my_table = soup.find('table',{'class':'wikitable'})
print(my_table)
```

```python
table = pd.read_html(str(my_table))
print(table)
type(table)
```

```python
df = pd.DataFrame(table[0])
print(df)
type(df)
```
